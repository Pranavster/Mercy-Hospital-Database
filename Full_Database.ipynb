```python
#0001
# Database Connection
def create_connection(host_name, user_name, user_password, db_name, port_no):
    connection = None
    try:
        connection = MySQLdb.connect(
            host=host_name,
            user=user_name,
            passwd=user_password,
            db=db_name,
            port=port_no
        )
        cursor = connection.cursor()
        print("Connection to MySQL DB successful")

    except OperationalError as err:
        if err.args[0] == 1049:
            print(f"Unknown database '{db_name}'")
            connection = MySQLdb.connect(
                host=host_name,
                user=user_name,
                passwd=user_password,
                db=db_name,
                port=port_no
            )
            cursor = connection.cursor()
            cursor.execute("Create Database{db_name}")
            cursor.select(f"{db_name}")
        else:
            print(f"The error '{err}' occurred")
            connection = None
    return connection
```

```python
#0002
#getting individual passwords to SQL server
def get_passwords(self):
    max_retries = 3
    retries = 0
    while retries < max_retries:
        try:
            password = input("Please enter password")
            database = input("Please enter database")
            connection = self.create_connection("localhost","root",password, database,3306)
        except MySQLdb.Error as err: # error checking mechanism
            if err.args[0] == 1205:
                retries += 1
                print(f"Lock wait timeout error. Retrying ({retries}/{max_retries})...")
                time.sleep(1)
            else:
                raise
        break
    else:
        print("Max retries reached. Transaction failed.")
```

```python
#0003
#Reading queries with result.       
def execute_read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        return result
    except OperationalError as e:
        print(f"The error '{e}' occurred")
```

```python
#0004
#For queries that have no return, 
def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query executed successfully")
    except OperationalError as e:
        print(f"The error '{e}' occurred")
```

```python
# OLAP Functions
# Group BY
# Query Name: Patient Count By Doctor
def group_by_patient_per_doctors(connection):
    query = """
    SELECT Doctor_ID, COUNT(*) AS doctor_count
    FROM Patient
    GROUP BY Doctor_ID
    ORDER BY Doctor_ID;
    """
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        results = cursor.fetchall()
        results_text="Patient Count by Doctor:\n"
        print("Patient Count by Doctor:")
        Total_docs =0
        for result in results:
            Doctor_ID, doctor_count = result
            Doctor_ID = Doctor_ID if Doctor_ID is not None else "Unassigned"
            results_text+= (str(Doctor_ID)+"    ")
            doctor_count = doctor_count if doctor_count is not None else ""
            Total_docs +=doctor_count
            results_text+= (str(doctor_count)+"\n")
            print(f"{Doctor_ID:<50}{doctor_count:<10}")
        results_text += ("Total=  "+str(Total_docs))
        return results_text
    except OperationalError as e:
        print(f"The error '{e}' occurred")

```

```python
# OLAP Functions
# Group BY
# Query Name: Doctor Count By Department
def group_by_doctors_per_dept(connection):
    query = """
    SELECT dept_id, COUNT(*) AS doctor_count
    FROM doctors
    GROUP BY dept_id
    ORDER BY dept_id;
    """
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        results = cursor.fetchall()
        results_text="Doctor Count by Department:\n"
        print("Doctor Count by Department:")
        Total_docs =0
        for result in results:
            dept_id, doctor_count = result
            dept_id = dept_id if dept_id is not None else "Unassigned"
            results_text+= (str(dept_id)+"    ")
            doctor_count = doctor_count if doctor_count is not None else ""
            Total_docs +=doctor_count
            results_text+= (str(doctor_count)+"\n")
            print(f"{dept_id:<50}{doctor_count:<10}")
        results_text += ("Total=  "+str(Total_docs))
        return results_text
    except OperationalError as e:
        print(f"The error '{e}' occurred")

```

```python
# ROLLUP
# Query Name: Procedure Count By Doctor with ROLLUP
def rollup_procedures_by_doctor(connection):
    # This query will count the procedures per doctor and include a rollup total.
    query = """
    SELECT s.Doctor_ID, COUNT(s.Procedure_Code) AS Procedure_Count
    FROM schedule s
    JOIN proceduretype p ON s.Procedure_Code = p.Procedure_Code
    GROUP BY s.Doctor_ID WITH ROLLUP;
    """
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        results = cursor.fetchall()
        print("Procedure Count by Doctor with Rollup:")
        results_text="Procedure Count by Doctor with Rollup:\n"
        for result in results:
            doctor_id, procedure_count = result
            
           
            if doctor_id is None:
                print(f"{'Total':<50}{procedure_count:<10}")
            else:
                doctor_id = doctor_id  
                results_text+= (str(doctor_id)+"    ")
                procedure_count = procedure_count if procedure_count is not None else "None"
                results_text+= (str(procedure_count)+"\n")
                print(f"{doctor_id:<50}{procedure_count:<10}")
        return results_text
    except OperationalError as e:
        print(f"The error '{e}' occurred")

```

```python
# CUBE
# Query Name: Procedure Count By Doctor with CUBE
def simulated_cube_procedures_by_doctor(connection):
    # This function simulates the CUBE operation in MySQL by manually unioning the different levels of aggregation.
    cube_simulation_query = """
    SELECT s.Doctor_ID, COUNT(s.Procedure_Code) AS Procedure_Count
    FROM schedule s
    JOIN proceduretype p ON s.Procedure_Code = p.Procedure_Code
    GROUP BY s.Doctor_ID
    UNION ALL
    SELECT NULL as Doctor_ID, COUNT(s.Procedure_Code) AS Procedure_Count
    FROM schedule s
    JOIN proceduretype p ON s.Procedure_Code = p.Procedure_Code;
    """
    cursor = connection.cursor()
    try:
        cursor.execute(cube_simulation_query)
        results = cursor.fetchall()
        print("Simulated Cube - Procedure Count by Doctor:")
        results_text="Procedure Count by Doctor by Cube:\n"
        
        for result in results:
            doctor_id, procedure_count = result
            doctor_id = doctor_id if doctor_id is not None else "Total"
            results_text+= (str(doctor_id)+"    ")
            procedure_count = procedure_count if procedure_count is not None else "None"
            results_text+= (str(procedure_count)+"\n")
            print(f"{doctor_id:<50}{procedure_count:<10}")
        return results_text
    except OperationalError as e:
        print(f"The error '{e}' occurred")
```
